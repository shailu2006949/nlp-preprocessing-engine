# **Task 1  Conceptual Understanding**



1. What is the difference between "Love" and "love" in NLP?

In NLP, "Love" and "love" are treated as two different tokens because NLP systems are case-sensitive. This unnecessarily increases the vocabulary size and may affect model performance. Hence, text is usually converted to lowercase during preprocessing.

2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many unnecessary words that do not carry significant meaning. This increases noise, memory usage, and can reduce the performance of machine learning models.

3. Provide two real-world scenarios where removing stopwords can be harmful.

In sentiment analysis, removing words like "not" can change the meaning of a sentence (e.g., "not good" becomes "good").
In question-answering systems or chatbots, words like "what", "how", and "where" are important to understand the query.

4. Explain the difference between stemming and lemmatization.

Stemming reduces words to their root form by cutting off letters (e.g., "running" → "runn"), which may not be a valid word. Lemmatization converts words to their meaningful base form using vocabulary and grammar rules (e.g., "running" → "run").

# **Task 2 Advanced Preprocessing Function**

```python
def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis & special chars
    text = re.sub(r'[^\w\s]', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = word_tokenize(text)

    # Remove very short tokens (≤2) except no, not
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

# **Task 3 Stress Testing (10 sentences)**

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:
    tokens, clean = preprocess_text(s)

    print("Original Text :", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-"*60)

    results.append(tokens)
```

# **Task 4 Token Analytics**

```python
for i, tokens in enumerate(results):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(t) for t in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-"*50)
```

**Analysis Answers**

The sentence with the most noise is: "Get 100% FREE access now!!!" because it contains numbers, symbols, and capital letters.

The sentence that retained the most meaningful tokens after cleaning is: "I am not happy with this" because it preserves important sentiment words like "not" and "happy".

# **Task 5 Frequency Analysis**

```python
from collections import Counter

# Combine all tokens from Task 3 results
all_tokens = [token for sublist in results for token in sublist]

freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(freq.most_common()[-5:])
```

# **Task 6 Full Pipeline**

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test the pipeline with previous sentences
output = full_pipeline(sentences)
output
```
