**Use the SIMPLEST chatbot method (pipeline)**

```python
from transformers import pipeline

chatbot = pipeline(
    "text-generation",
    model="microsoft/DialoGPT-medium"
)
```

**Chat function**

```python
def ask_bot(user_text):
    prompt = f"User: {user_text}\nBot:"
    output = chatbot(prompt, max_length=100, pad_token_id=50256)
    reply = output[0]["generated_text"].split("Bot:")[-1]
    return reply.strip()
```

**Interactive chatbot**

```python
print("Chatbot ready! Type 'exit' to stop.\n")

while True:
    user = input("You: ")
    if user.lower() == "exit":
        print("Bot: Goodbye!")
        break

    response = ask_bot(user)
    print("Bot:", response)
```
